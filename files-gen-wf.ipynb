This is a code to generate a file for a desired GROMACS md. The main goal is to generate 2 script files:
- The `steps.sh` file, which is the one which would be ran locally to prepare the system
- The `-script.sh` file, which is the one which would be ran in ACENET remotely, and that would already have the general format of both the files generated before, and the files that will be generated.

Its a rough script, been worked on but with the purpose of making my own experiments more quickly, and with a general structure to analyze parameters vs outcome.

```python
import colfib as cf

name_output = f"coltest"

# Generate folder with files for preparation
cf.generate_files_prep(name = name_output,
                        pdb_file = f"colmol_cutd.pdb",
                        ff = "amber99sb-star-ildnp",
                        watertype = "tip3p",
                        dimensions = None,
                        angles = None,
                        distance = "3.0",
                        solvent_config = "spc216.gro",
                        pion = "NA",
                        nion = "CL",
                        ccon = None,
                        emtol = "250.0",
                        chain_org = True)
```

Most of the parameters are self explanatory, with the exception of:
- `distance`: Is used instead of the parameters `dimensions` and `angles` to, instead of building a custom box, build a box based on gap between protein and edges. By default is `None`.
- `ccon`: Used to specify if an extra cation solution concentration wants to be added. By default is `None`.
- `chain_org`: To make a folder of the itp files generated (in case there are too much files, such as with the collagen fibril). By default is `True`.

For executing the preparation one then should go to the folder `-files`, and inside allow the script execute as:
- `chmod +x ./steps.sh`
- `./steps.sh`

Having the correct gromacs verision (the one generated in the script) sourced for `gmx`.

This part has the stuff related with the preparation of the other parts of the process (the EM, equilibration, and MD). Its super rough as its just useful for me, but may be worked on more in the future for better costumization.

```python
# Generate script and .mdp files for Equilibration and MD
cf.generate_script(name = name_output,
                     runtime = "00:30:00",
                     user = "a00836485@tec.mx",
                     steps = [1,1,1,1])

cf.mdp_parms(name = name_output,
               cutoff = 1.2,
               temperature = "300.0",
               pressure = "1.0",
               taup = "15.0",
               nvt_time = 1,
               npt_time = 5,
               md_time = 25,
               dt_eq = 1,
               dt_md = 2)
```

The units are:
- Temperature: Kelvin
- Pressure: Bar
- Time: Nanosecods
- DeltaTime: Femtoseconds
